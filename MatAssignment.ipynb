```python
Q1 -> What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans -> Matplotlib is a popular data visualization library in Python that provides a wide range of tools for creating static, animated, and interactive visualizations in Python. It is widely used for creating graphs, charts, histograms, and other types of visualizations in data analysis, scientific research, and machine learning projects.

Matplotlib is used to create high-quality 2D and 3D visualizations in Python, and it is widely supported in the Python community. It provides a variety of visualization tools and customization options, and it can be integrated with other Python libraries like NumPy, Pandas, and Seaborn.

The Pyplot module of Matplotlib is a collection of functions that provide a convenient interface for creating various types of plots. Some of the plots that can be created using the Pyplot module of Matplotlib include:

1) Line plot: A simple line plot can be created using the Pyplot module by specifying the x and y values and using the plot() function.

2) Scatter plot: A scatter plot can be created using the scatter() function in the Pyplot module by specifying the x and y values and customizing the marker size and color.

3) Bar plot: A bar plot can be created using the bar() function in the Pyplot module by specifying the x and y values and customizing the color and width of the bars.

4) Histogram: A histogram can be created using the hist() function in the Pyplot module by specifying the data and customizing the bin size and color.

5) Box plot: A box plot can be created using the boxplot() function in the Pyplot module by specifying the data and customizing the box and whisker styles.
```

```python
Q2 ->What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Ans -> import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))

plt.scatter(x, y)
plt.title("seed graph")
plt.xlabel("x-axis")
plt.ylabel("y axis")
plt.show()
```

```python
Q4 -> What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Ans ->  A bar plot is a visualization tool used to represent categorical data with rectangular bars. Bar plots are useful when we want to compare the quantities of different categories or groups. The height or length of the bars represents the values of the data, and the categories are displayed on the x-axis or y-axis.

Bar plots can be used to display various types of data, such as counts, frequencies, proportions, and averages.

To plot a bar plot and a horizontal bar plot using the given data, we can use the bar() and barh() functions of the Pyplot module of Matplotlib. Here is the code to plot both:
    
import matplotlib.pyplot as plt 
import numpy

company = np.array(['Apple' , 'Microsoft' , 'Google' , 'AMD'])
profit = np.array([3000 , 8000 , 1000 , 100000])

plt.bar(company , profit)
plt.title('company_profit')
plt.xlabel('company')
plt.ylabel('profit')
plt.show()


# on horizontal 

plt.barh(company , profit)
plt.title('company profit')
plt.xlabel('profit')
plt.ylabel('company')
plt.show()
```

```python
Q5 -> What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Ans -> A box plot is a visualization tool that displays the distribution of a dataset by showing the median, quartiles, and outliers. The box in the plot represents the interquartile range (IQR) which is the range between the first quartile (Q1) and the third quartile (Q3), and the line inside the box represents the median. The whiskers extend from the box to the lowest and highest data points within 1.5 times the IQR, and any data points outside of the whiskers are considered outliers.

Box plots are useful when we want to compare the distribution of different datasets or to identify any outliers in the data.

To plot a box plot using the given data, we can use the boxplot() function of the Pyplot module of Matplotlib. Here is the code to plot the box plot:
    
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100 , 10 ,200)
box2 = np.random.normal(90 , 20 , 200)
data = [box1 , box2]

plt.boxplot(data)
plt.title("box plot")
plt.xticks([1,2] , ['Box 1' , 'Box 2'])
plt.ylabel("value")
plt.show()
```
